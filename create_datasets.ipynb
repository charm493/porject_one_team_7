```python
import pandas as pd
from atwater_functions import load_data, merge_data, plot_with_dividers, calculate_election_period_prices, generate_lame_duck_dataset
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from datetime import timedelta
```

```python
energy_file_path = 'raw_data/Energy Select Sector SPDR Fund.csv'
consumer_staples_file_path = 'raw_data/Consumer Staples Select Sector SPDR Fund.csv'
financial_file_path = 'raw_data/Financial Select Sector SPDR Fund.csv'
health_care_file_path = 'raw_data/Health Care Select Sector SPDR Fund.csv'
tech_file_path = 'raw_data/Technology Select Sector SPDR Fund.csv'
political_file_path = 'raw_data/political_control_final.csv'
utilities_file_path = 'raw_data/Utilities Select Sector SPDR Fund.csv'
```

```python
# Create energy dataset

# Load data
stock_file_path = energy_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
energy_data = merge_data(stock_data, political_data)
energy_data['Sector'] = "Energy"

energy_data = energy_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
energy_data = energy_data.drop(columns=['7. dividend amount'])
energy_data.to_csv('cleaned/energy_data_merged.csv', index=False)
energy_data
```

```python
# Create Consumer Staples dataset

# Load data
stock_file_path = consumer_staples_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
consumer_staples_data = merge_data(stock_data, political_data)
consumer_staples_data['Sector'] = "Consumer Staples"

consumer_staples_data = consumer_staples_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
consumer_staples_data = consumer_staples_data.drop(columns=['7. dividend amount'])
consumer_staples_data.to_csv('cleaned/consumer_staples_data_merged.csv', index=False)
consumer_staples_data
```

```python
# Create Financial dataset

# Load data
stock_file_path = financial_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
financial_data = merge_data(stock_data, political_data)
financial_data['Sector'] = "Financial"

financial_data = financial_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
financial_data = financial_data.drop(columns=['7. dividend amount'])
financial_data.to_csv('cleaned/financial_data_merged.csv', index=False)
financial_data
```

```python
# Create health care dataset

# Load data
stock_file_path = health_care_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
health_care_data = merge_data(stock_data, political_data)
health_care_data['Sector'] = "Health Care"

health_care_data = health_care_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
health_care_data = health_care_data.drop(columns=['7. dividend amount'])
health_care_data.to_csv('cleaned/health_care_data_merged.csv', index=False)
health_care_data
```

```python
# Create technology dataset

# Load data
stock_file_path = tech_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
tech_data = merge_data(stock_data, political_data)
tech_data['Sector'] = "Technology"

tech_data = tech_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
tech_data = tech_data.drop(columns=['7. dividend amount'])
tech_data.to_csv('cleaned/technology_data_merged.csv', index=False)
tech_data
```

```python
# Create utilities dataset

# Load data
stock_file_path = utilities_file_path
stock_data, political_data = load_data(stock_file_path, political_file_path)

# Merge datasets
utilities_data = merge_data(stock_data, political_data)
utilities_data['Sector'] = "Utilities"

utilities_data = utilities_data.rename(
    columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",	
        "4. close": "Close",	
        "5. adjusted close": "Adjusted Close",
        "6. volume": "Volume"

    }
)
utilities_data = utilities_data.drop(columns=['7. dividend amount'])
utilities_data.to_csv('cleaned/utilities_data_merged.csv', index=False)
utilities_data
```

```python
combined_cleaned_data = pd.concat(
    [energy_data, tech_data, utilities_data, health_care_data, consumer_staples_data, financial_data],
    ignore_index=True
)
combined_cleaned_data.to_csv('cleaned/combined_merged_dataset.csv', index=False)
```

```python
# Create percent change dataframes

consumer_staples_percent_change = calculate_election_period_prices(consumer_staples_data, output_csv_path='percent_change/consumer_staples_percent_change_per_term.csv')

energy_percent_change = calculate_election_period_prices(energy_data, output_csv_path='percent_change/energy_percent_change_per_term.csv')

financial_percent_change = calculate_election_period_prices(financial_data, output_csv_path='percent_change/financial_percent_change_per_term.csv')

health_care_percent_change = calculate_election_period_prices(health_care_data, output_csv_path='percent_change/health_care_percent_change_per_term.csv')

tech_percent_change = calculate_election_period_prices(tech_data, output_csv_path='percent_change/technology_percent_change_per_term.csv')

utilities_percent_change = calculate_election_period_prices(utilities_data, output_csv_path='percent_change/utilities_percent_change_per_term.csv')

combined_percent_change = pd.concat(
    [consumer_staples_percent_change, energy_percent_change, financial_percent_change, health_care_percent_change, tech_percent_change, utilities_percent_change],
    ignore_index=True
)

# Combine sectors into one percent change dataframe

combined_percent_change.to_csv('percent_change/combined_percent_change_data.csv', index=False)
combined_percent_change
```

```python
# Create lame duck datasets
energy_lame_duck = generate_lame_duck_dataset(energy_data, output_csv_path='lame_duck/energy_lame_duck_data.csv')
consumer_staples_lame_duck = generate_lame_duck_dataset(consumer_staples_data, output_csv_path='lame_duck/consumer_staples_lame_duck_data.csv')
financial_lame_duck = generate_lame_duck_dataset(financial_data, output_csv_path='lame_duck/financial_lame_duck_data.csv')
health_care_lame_duck = generate_lame_duck_dataset(health_care_data, output_csv_path='lame_duck/health_care_lame_duck_data.csv')
tech_lame_duck = generate_lame_duck_dataset(tech_data, output_csv_path='lame_duck/technology_lame_duck_data.csv')
utilities_lame_duck = generate_lame_duck_dataset(utilities_data, output_csv_path='lame_duck/utilities_lame_duck_data.csv')

# Combine and write to csv
combined_lame_duck = pd.concat(
    [energy_lame_duck, consumer_staples_lame_duck, financial_lame_duck, health_care_lame_duck, tech_lame_duck, utilities_lame_duck],
    ignore_index=True
)
combined_lame_duck.to_csv('lame_duck/combined_lame_duck_data.csv', index=False)
```

```python
# create lame duck percent changes

energy_ld_perc_change = calculate_election_period_prices(energy_lame_duck, output_csv_path='lame_duck_perc_change/energy_ld_perc_change.csv')
health_care_ld_perc_change = calculate_election_period_prices(health_care_lame_duck, output_csv_path='lame_duck_perc_change/health_care_ld_perc_change.csv')
cons_staples_ld_perc_change = calculate_election_period_prices(consumer_staples_lame_duck, output_csv_path='lame_duck_perc_change/cons_staples_ld_perc_change.csv')
financial_ld_perc_change = calculate_election_period_prices(financial_lame_duck, output_csv_path='lame_duck_perc_change/financial_ld_perc_change.csv')
tech_ld_perc_change = calculate_election_period_prices(tech_lame_duck, output_csv_path='lame_duck_perc_change/technology_ld_perc_change.csv')
utilities_ld_perc_change = calculate_election_period_prices(utilities_lame_duck, output_csv_path='lame_duck_perc_change/utilities_ld_perc_change.csv')

#combined lame duck percent changes

combined_lame_duck_perc_change = pd.concat(
    [energy_ld_perc_change, health_care_ld_perc_change, cons_staples_ld_perc_change, financial_ld_perc_change, tech_ld_perc_change, utilities_ld_perc_change],
    ignore_index=True
)
combined_lame_duck_perc_change.to_csv('lame_duck_perc_change/combined_lame_duck_perc_change.csv', index=False)
```

```python
# View plots by changing percent change input df

input_df = energy_ld_perc_change
# Extract relevant columns for the bar plot
plot_data = input_df[['End of Term', 'Percent Change']]

# Convert 'End of Term' to datetime for better sorting and plotting
plot_data['End of Term'] = pd.to_datetime(plot_data['End of Term'])

# Sort by 'End of Term' to ensure chronological order
plot_data = plot_data.sort_values('End of Term')

# Assign colors based on the 'President' column
colors = input_df['President'].apply(lambda x: 'red' if x == 'Republican' else 'blue')


# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(plot_data['End of Term'].dt.strftime('%Y-%m-%d'), 
        plot_data['Percent Change'], 
        color=colors)

# Add legend
red_patch = mpatches.Patch(color='red', label='Republican President')
blue_patch = mpatches.Patch(color='blue', label='Democratic President')
plt.legend(handles=[red_patch, blue_patch], title='President', loc='upper left')

# Customize plot
plt.xticks(rotation=45, ha='right')
plt.title('Percent Change by Election Term')
plt.xlabel('End of Term Date')
plt.ylabel('Percent Change (%)')
plt.tight_layout()
plt.show()
```
