```python
#  forecast next two year stock performance
from prophet import Prophet
from datetime import datetime, timedelta


def forecast_stock_for_next_two_years(sector):

    clean_df = df.loc[(df["Sector"] == sector) & (df["President"] == "Republican")]
    clean_df = clean_df[["Date", "Close"]]
    clean_df.sort_values(by="Date", inplace=True, ascending=True)
    clean_df = clean_df.reset_index()
    clean_df.drop('index', axis=1, inplace=True)

    clean_df = clean_df.rename(columns={"Date":"ds", "Close": "y"})
    model = Prophet()
    model.fit(clean_df)
    
    today = datetime.today()
    last_date = pd.to_datetime(clean_df['ds']).max()  # Last date in my dataset
    days_to_extend = (today - last_date).days  # Gap between the last date and today

    
    pre_df = model.make_future_dataframe(periods=730 + days_to_extend, freq="D")
    pre_df = model.predict(pre_df)


    fig = model.plot(pre_df);
    ax = fig.gca()
    ax.set_title(f"{sector} stock Forecast")

for sector in df["Sector"].unique():
    forecast_stock_for_next_two_years(sector)
```
