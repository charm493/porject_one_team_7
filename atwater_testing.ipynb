```python
combined_sectors = combined_sectors.set_index("Date")
combined_sectors
```

```python
combined_sectors = combined_sectors.drop(columns=['1. open', '2. high', '3. low', '5. adjusted close', '7. dividend amount'])
combined_sectors
```

```python
combined_sectors = combined_sectors.rename(columns={
    "4. close": "Close",
    "6. volume": "Volume"
})
combined_sectors
```

```python
combined_sectors = combined_sectors.sort_values(by="Date", ascending=False)
combined_sectors
```

```python
# Political Scenarios
uniques_df = combined_sectors.copy().reset_index()
columns = ['President', 'House', 'Senate']
uniques_df = uniques_df[columns].drop_duplicates()
uniques_df.reset_index(drop=True)
uniques_df
```

```python
# lame duck scenario
# Filter for dates after November 10 and before January 20 of any year
combined_sectors = combined_sectors.reset_index()
lame_duck_df = combined_sectors[
    (combined_sectors['Date'].dt.month == 11) & (combined_sectors['Date'].dt.day > 10) |
    (combined_sectors['Date'].dt.month == 12) |
    (combined_sectors['Date'].dt.month == 1) & (combined_sectors['Date'].dt.day < 20)
]
lame_duck_df
```

```python
# Plot the data as an area chart 
plt.figure(figsize=(12, 6)) 
plt.fill_between(lame_duck_df['Date'], lame_duck_df['Close'], color="skyblue", alpha=0.4) 
plt.plot(lame_duck_df['Date'], lame_duck_df['Close'], color="Slateblue", alpha=0.6, linewidth=2) 
plt.xlabel('Date') 
plt.ylabel('Close') 
plt.title('Change Over Time') 
plt.grid(True) 
plt.show()
```

```python
group_df = lame_duck_df.groupby("Sector")["Close"].mean()
group_df.plot.bar()
```

```python
lame_duck_df.loc[lame_duck_df["Sector"] == "Energy"]["Close"].pct_change()
group_df = lame_duck_df.groupby("Sector")["Close"].pct_change()
group_df.plot()
```

```python


grouping_close = lame_duck_df[lame_duck_df["Sector"] == "Energy"].groupby(by=["Start Date", "End Date", "President"]).mean("Close")
grouping_close = grouping_close.reset_index()
plt.plot(grouping_close['Start Date'], grouping_close['Close'])

for i, j in grouping_close.iterrows():
    plt.axvline(x=pd.Timestamp(j["Start Date"]), color=("blue" if j["President"] == "Democratic" else "red"), linestyle='--', label='Event Date')
plt.show()

```

```python
grouping = lame_duck_df[lame_duck_df["Sector"] == "Energy"].groupby(by=["Start Date", "End Date", "President"]).mean("Volume")
grouping = grouping.reset_index()
grouping
plt.plot(grouping['Start Date'], grouping['Volume'])
for i, j in grouping.iterrows():
    plt.axvline(x=pd.Timestamp(j["Start Date"]), color=("blue" if j["President"] == "Democratic" else "red"), linestyle='--', label='Event Date')
plt.show()
```
