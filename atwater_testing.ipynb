```python
combined_sectors = combined_sectors.set_index("Date")
combined_sectors
```

```python
combined_sectors = combined_sectors.drop(columns=['1. open', '2. high', '3. low', '5. adjusted close', '7. dividend amount'])
combined_sectors
```

```python
combined_sectors = combined_sectors.rename(columns={
    "4. close": "Close",
    "6. volume": "Volume"
})
combined_sectors
```

```python
combined_sectors = combined_sectors.sort_values(by="Date", ascending=False)
combined_sectors
```

```python
# Political Scenarios
uniques_df = combined_sectors.copy().reset_index()
columns = ['President', 'House', 'Senate']
uniques_df = uniques_df[columns].drop_duplicates()
uniques_df.reset_index(drop=True)
uniques_df
```

```python
# lame duck scenario
# Filter for dates after November 10 and before January 20 of any year
combined_sectors = combined_sectors.reset_index()
lame_duck_df = combined_sectors[
    (combined_sectors['Date'].dt.month == 11) & (combined_sectors['Date'].dt.day > 10) |
    (combined_sectors['Date'].dt.month == 12) |
    (combined_sectors['Date'].dt.month == 1) & (combined_sectors['Date'].dt.day < 20)
]
lame_duck_df
```

```python
lame_duck_energy = lame_duck_df[lame_duck_df['Sector']=='Energy']
lame_duck_energy
```
